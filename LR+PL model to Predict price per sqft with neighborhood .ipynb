```python
df = pd.read_csv(r"C:\Users\Nassoh\Desktop\DataSienceMyProject\Dubai Houses analize\Data\Dubai.csv")
print(df.shape)
df.head()   
```

# Spilt

```python
target = "price"
features = ["neighborhood"]
y_train =df[target]
X_train =df[features]
```

# Baseline

```python
y_mean =y_train.mean()
y_mean
y_pred_baseline =[y_mean]*len(y_train)
len(y_pred_baseline)
mae_baseline =mean_absolute_error(y_train,y_pred_baseline)
print("Mean apt price", round(y_mean, 2))
print("Baseline MAE:", round(mae_baseline, 2))
```

# Build Model

```python
model =make_pipeline( OneHotEncoder(use_cat_names=True),Ridge())
model.fit(X_train,y_train)
```

# Evaluate

```python
y_pred_training =model.predict(X_train)
mae_training =mean_absolute_error(y_train,y_pred_training)
print("Training MAE:", round(mae_training, 2))
print("Baseline MAE:", round(mae_baseline, 2))
```

# Intercept,coefficient and Feature names For Model

```python
intercept =model.named_steps['ridge'].intercept_
coefficients =model.named_steps['ridge'].coef_
feature_names =model.named_steps['onehotencoder'].get_feature_names()
```

# Feature names and equation For Model

```python
feat_imp =pd.Series(coefficients,index=feature_names)
feat_imp
print(f"price = {intercept.round(2)}")
for f, c in feat_imp.items():
    print(f"+ ({round(c, 2)} * {f})")
```
